### 3. Antecedentes

#### 3.0 Conocimiento Situado

Desde las teorías feministas de finales del siglo XX se ha cuestionado la idea de un conocimiento universal y neutro, subrayando que toda producción de saber está situada en cuerpos, posiciones sociales y relaciones de poder (Haraway, 1988; Harding, 1991). En esta línea, la investigación muestra que quién investiga —sus disposiciones, tiempos, recursos y vínculos— afecta qué preguntas formula y cómo valida sus resultados (Collins, 1990; Muñoz, año). Así, género y clase no son atributos colaterales sino condiciones de posibilidad de la indagación: ordenan trayectorias, distribuyen oportunidades y modelan agendas de conocimiento (Bourdieu, 1984, 1986).

Este sustrato teórico ayuda a entender por qué la estructura de género incide en la selección de carreras, la persistencia y las prácticas dentro de campos disciplinares. La literatura documenta una mayor presencia de mujeres en Ciencias Sociales y Humanidades y su subrepresentación en STEM, relación mediada por estereotipos de habilidad y normas culturales que se reproducen institucionalmente (Xie & Shauman, 2003; Ceci & Williams, 2011; Moss-Racusin et al., 2012). Tales mecanismos no sólo afectan los accesos; también estratifican la agencia epistémica, orientando las preguntas que se consideran legítimas y los marcos que ganan prestigio.

A su vez, las desigualdades de clase operan en la provisión diferencial de capitales —económico, social y cultural— que condicionan el acceso a universidades de alto prestigio, la calidad de los vínculos académicos y la productividad científica (Bourdieu, 1986; Merton, 1968). En mercados académicos jerárquicos, las instituciones de élite tienden a contratar a egresados de instituciones afines, reproduciendo el cierre social y consolidando circuitos reputacionales (Clauset, Arbesman & Larremore, 2015). Este patrón se traduce en condiciones asimétricas de producción (recursos, ayudantías, redes internacionales) y ventajas acumulativas que sesgan la evaluación de méritos y la visibilidad de resultados.

Bajo este marco, el estado del arte exhibe esfuerzos parciales por mapear empíricamente la relación entre género, clase y temáticas de investigación; sin embargo, las aproximaciones que vinculan características sociales de investigadores con huellas textuales (p. ej., tópicos en resúmenes y publicaciones) siguen siendo escasas y fragmentarias. Esta tesis propone avanzar un paso más: identificar tópicos predominantes a nivel de autor y estudiarlos como expresiones situadas de agencia en campos disciplinares, incorporando explícitamente la clase social de origen y el género. El foco empírico —la cohorte de Becarios/as ANID 2014–2016— permite indagar cómo estas dimensiones inciden en el posicionamiento institucional, en la productividad y en las preguntas de investigación que efectivamente se formulan.

#### 3.1. Desigualdades estructurales en educación superior (Chile y comparados)

La educación superior se ha expandido en cobertura, pero no por ello ha desarticulado las lógicas de reproducción social que estratifican el acceso, la permanencia y la consecución de posiciones académicas. En términos comparados, existe un amplio consenso en que la selección universitaria y la posterior inserción laboral en la academia operan mediante mecanismos de clase, género y origen socio-institucional que acumulan ventajas y desventajas en el tiempo (Moss-Racusin, Dovidio, Brescoll, Graham & Handelsman, 2012, pp. 16474–16476; BCN, 2023, pp. 12–15). En Chile, el patrón es particularmente visible por la confluencia de tres procesos: (i) masificación con fuerte segmentación institucional, (ii) financiamiento público focalizado que incentiva trayectorias específicas de posgrado (p. ej., Becas Chile), y (iii) densificación de credenciales que no siempre se traduce en mejores posiciones en el campo científico local (DIPRES, 2024, pp. 23–27; BCN, 2023, pp. 18–21). Estos procesos se retroalimentan.

En esta clave, la desigualdad no se entiende sólo como “diferencia” sino como estructura de oportunidades que asigna probabilidades desiguales de tránsito por la carrera académica. Un cuerpo de evidencia experimental ha mostrado sesgos sistemáticos de género en evaluaciones “ciegas” de méritos—por ejemplo, en cartas de recomendación, valoración de CV y evaluación de competencias percibidas—con efectos medibles en la contratación y asignación salarial (Moss-Racusin et al., 2012, pp. 16476–16479). En contextos periféricos, estas asimetrías se combinan con dependencia epistémica y con incentivos de publicación que privilegian ciertos idiomas, revistas y estilos de investigación, profundizando brechas entre instituciones y personas (Global Qualitative Sociology, 2024, s. p.). El resultado es un campo académico jerárquico donde la procedencia (de clase y de institución) pesa tanto como la productividad observable (DIPRES, 2024, pp. 26–27).

#### 3.2. Modelo androcéntrico y sesgos en STEM; dependencia epistémica en Ciencias Sociales

La literatura internacional en STEM ha documentado un modelo androcéntrico que opera desde la socialización temprana hasta los procesos de contratación y promoción: los hombres son evaluados como más “competentes” y “contratables” incluso a igualdad de méritos, con brechas salariales iniciales (Moss-Racusin et al., 2012, pp. 16476–16478). Este patrón se entrelaza con sesgos de idioma y centro/periferia: publicar en inglés, en revistas de alto impacto y desde instituciones “core” multiplica la visibilidad y las citaciones, mientras que la producción desde el Sur Global enfrenta mayores costos de legitimación (Global Qualitative Sociology, 2024, s. p.). En Ciencias Sociales, esto se expresa como dependencia epistémica: marcos teóricos, agendas y criterios de validación se importan desde centros hegemónicos, configurando asimetrías de reconocimiento (ERIC, 2024, pp. 205–209; Tuning Journal, 2020, pp. 45–47).

Chile no es ajeno a esta dinámica. La orientación de políticas públicas hacia la internacionalización del posgrado—con instrumentos como Becas Chile—buscó elevar estándares y redes, pero también generó tensiones de retorno, reinserción y pertinencia de competencias respecto de la estructura local de oportunidades (BCN, 2023, pp. 22–28). Allí donde los sesgos de género y de origen institucional persisten, la inserción posdoctoral y el acceso a plazas permanentes pueden volverse cuellos de botella, incluso para trayectorias con alta productividad (DIPRES, 2024, pp. 24–26).

#### 3.3. Evidencia nacional sobre estratificación por clase/género y expansión doctoral

La expansión doctoral chilena de las dos últimas décadas elevó el número de egresados y retornados, pero no homogeneizó su inserción académica. Los informes de evaluación del programa Becas Chile reportan heterogeneidad en el cumplimiento de obligaciones de retorno, tasas de reinserción y correspondencia entre formación e inserción (BCN, 2023, pp. 8–11). Simultáneamente, el monitoreo presupuestario de DIPRES identifica desafíos en la articulación entre inversión pública y efectos en el sistema de ciencia, tecnología e innovación, con brechas por área, institución de origen y género (DIPRES, 2024, pp. 19–27). Estos hallazgos se alinean con la evidencia comparada que muestra cómo los mecanismos de cierre social—acreditación programática, prestigio institucional, redes supervisor-estudiante—operan como filtros de acceso a posiciones (ERIC, 2024, pp. 210–213).

Un punto crucial para esta tesis es que la producción científica no es neutra: las elecciones de temas, métodos y foros de publicación codifican pertenencias de campo y subcampo, y esas pertenencias median la circulación de reconocimiento y oportunidades (PNAS, 2012, pp. 16476–16479). Así, estudiar tópicos y su distribución entre autores e instituciones no es sólo un ejercicio bibliométrico, sino una forma de trazar la ecología de oportunidades que enmarca las trayectorias (Tuning Journal, 2020, pp. 40–44).

#### 3.4. Relevancia de grafos de conocimiento para trayectorias académicas

Los grafos de conocimiento (KG) permiten integrar entidades y relaciones en una estructura formal sobre la cual razonar y predecir (Xie, Zhang & Yuan, 2022, pp. 1–3). En el plano metodológico, un KG académico facilita consultas complejas y tareas de inferencia que son imposibles o costosas en tablas relacionales tradicionales (Springer, 2019, pp. 127–131).

En términos de política científica, esta arquitectura dialoga directamente con los objetivos de programas públicos de formación avanzada: permite observar si la inversión se traduce en campos estratégicos, si las autoras y autores retornados se insertan en comunidades epistémicas afines y si persisten asimetrías de género/clase en ciertas áreas (DIPRES, 2024, pp. 25–27; BCN, 2023, pp. 22–28). Además, ancla empíricamente los énfasis de la formulación Fondecyt de esta investigación—esto es, el seguimiento de trayectorias y el análisis de mecanismos de estratificación—al proveer una infraestructura de datos e inferencia (Resumen formulación Fondecyt, 2024, s. p.).
